```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
```

```python
inputs = [-1, 0, 0.5, 1, 2]
```

```python
def linear_activation(x):
    return x

    linear_values = linear_activation(x)

# Linear Activation
linear_outputs = [linear_activation(x) for x in inputs]
print("Linear Activation:", linear_outputs)

plt.subplot(3, 2, 1)
plt.plot(x, linear_values, label='Linear Activation')
plt.title('Linear Activation')
plt.xlabel('x')
plt.ylabel('Activation')
plt.grid(True)
```

```python
def binary_unipolar_activation(x, threshold=0):
    return 1 if x >= threshold else 0

    binary_unipolar_values = binary_unipolar_activation(x)


binary_unipolar_outputs = [binary_unipolar_activation(x) for x in inputs]
print("Binary Unipolar Activation:", binary_unipolar_outputs)

plt.subplot(3, 2, 2)
plt.step(x, binary_unipolar_values, label='Binary Unipolar Activation', where='post')
plt.title('Binary Unipolar Activation')
plt.xlabel('x')
plt.ylabel('Activation')
plt.grid(True)
```

```python
def binary_bipolar_activation(x, threshold=0):
    return 1 if x >= threshold else -1

    binary_bipolar_values = binary_bipolar_activation(x)

# Binary Bipolar Activation

binary_bipolar_outputs = [binary_bipolar_activation(x) for x in inputs]
print("Binary Bipolar Activation:", binary_bipolar_outputs)

plt.subplot(3, 2, 3)
plt.step(x, binary_bipolar_values, label='Binary Bipolar Activation', where='post')
plt.title('Binary Bipolar Activation')
plt.xlabel('x')
plt.ylabel('Activation')
plt.grid(True)
```

```python
import numpy as np

def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))

    sigmoid_values = sigmoid_activation(x)

plt.subplot(3, 2, 4)
plt.plot(x, sigmoid_values, label='Sigmoid Activation')
plt.title('Sigmoid Activation')
plt.xlabel('x')
plt.ylabel('Activation')
plt.grid(True)
```

```python

def relu_activation(x):
    return max(0, x)
    relu_values = relu_activation(x)

#ReLU activation
relu_outputs = [relu_activation(x) for x in inputs]
print("ReLU Activation:", relu_outputs)

plt.subplot(3, 2, 5)
plt.plot(x, relu_values, label='ReLU Activation')
plt.title('ReLU Activation')
plt.xlabel('x')
plt.ylabel('Activation')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the range of input
x = np.arange(-10, 10.1, 0.1)
y_linear = x

# Plot Linear Activation
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y_linear, 'b', linewidth=2)
plt.title('Linear Activation')
plt.xlabel('x:input')
plt.ylabel('y')

# Plot Sigmoid Activation
y_sigmoid = 1 / (1 + np.exp(-x))
plt.subplot(2, 2, 2)
plt.plot(x, y_sigmoid, 'r', linewidth=2)
plt.title('Sigmoid Activation')
plt.xlabel('x:input')
plt.ylabel('y')

# Plot ReLU Activation
y_relu = np.maximum(0, x)
plt.subplot(2, 2, 3)
plt.plot(x, y_relu, 'g', linewidth=2)
plt.title('ReLU Activation')
plt.xlabel('x:input')
plt.ylabel('y')

# Plot Leaky ReLU Activation
y_lrelu = np.maximum(0.01 * x, x)
plt.subplot(2, 2, 4)
plt.plot(x, y_lrelu, 'y', linewidth=2)
plt.title('Leaky ReLU Activation')
plt.xlabel('x:input')
plt.ylabel('y')

# Display the plots
plt.tight_layout()
plt.show()
```
